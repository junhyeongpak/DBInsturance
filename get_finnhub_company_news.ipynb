```python
import finnhub
import os
from dotenv import load_dotenv
import pandas as pd
import time
```

```python
load_dotenv()

# Retrieve API key from .env
FINNHUB_API = os.getenv("FINNHUB_API_KEY")

print(FINNHUB_API)

# Setup client
finnhub_client = finnhub.Client(api_key=FINNHUB_API)
```

```python
# 다우존스 30 종목 티커 리스트
dow_30_tickers = ['BA', 'AAPL', 'UNH', 'GS', 'HD', 'MCD', 'V', 'MMM', 'MSFT', 
                  'CAT', 'DIS', 'JNJ', 'JPM', 'TRV', 'IBM', 'AXP', 'PG', 'CVX', 'WMT', 
                  'NKE', 'MRK', 'XOM', 'INTC', 'VZ', 'WBA', 'KO', 'DOW', 'CSCO', 'PFE']

# 날짜 범위 설정
start_date = "2024-02-20"
end_date = "2024-03-05"

# 뉴스 데이터를 저장할 리스트
all_news = []
```

```python
# 각 종목의 뉴스 가져오기
for ticker in dow_30_tickers:
    try:
        news = finnhub_client.company_news(ticker, _from=start_date, to=end_date)
        
        # 뉴스가 있는 경우 데이터 저장
        if news:
            for item in news:
                all_news.append({
                    "ticker": ticker,
                    "datetime": item.get("datetime"),
                    "headline": item.get("headline"),
                    "source": item.get("source"),
                    "summary": item.get("summary"),
                    "url": item.get("url")
                })
        
        # API 호출 제한을 피하기 위해 딜레이 추가 (필요 시 조절)
        time.sleep(1)
    except Exception as e:
        print(f"Error fetching news for {ticker}: {e}")
```

```python
# 데이터프레임 변환
df = pd.DataFrame(all_news)

# CSV 파일로 저장
df.to_csv("dow_30_news.csv", index=False, encoding="utf-8-sig")

# 데이터 확인
df.head()
```
